### :rabbit: 그래프 최소 비용 문제

- 최소 신장 트리 문제
    - 모든 정점들을 연결하는 간선들의 가중치 합 최소
    - 무향그래프: n개의 정점과 n-1개의 간선으로 구성
    1. 프림(Prim) 알고리즘 : 한 정점에 연결된 간선들 중 하나씩 선택하면서 최소 신장 트리를 만들어 가는 방식
    2. 크루스칼(Kruskal) 알고리즘 : 최소 가중치 간선을 하나씩 선택해서 최소 신장 트리를 찾는 방식
    
---
- 최단 경로 문제
    - 유향그래프에서 두 정점 사이의 경로들 중 간선의 가중치 합 최소
    - 단일 시작점 최단 경로 문제 : 출발점 -> 다른 모든 정점
        1. 다익스트라 : 음의 가중치 허용 X
        2. 벨만 포드(Bellman-Ford) : 음의 가중치 허용, 가중치 합이 음인 사이클 허용 X
    - 모든 쌍 최단 경로 문제 : 모든 정점 쌍 간의 최단 경로
        1. 플로이드-워샬 알고리즘
