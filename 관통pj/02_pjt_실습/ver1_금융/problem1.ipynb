## B. 데이터 전처리 - 2021년 이후의 종가 데이터 출력하기

```python
# CSV 파일 경로
csv_path = 'NFLX.csv'

# CSV 파일 읽어오기 (첫 번째, 마지막 열 제외)
df = pd.read_csv(csv_path, usecols=range(0, 5))

# 필터링 가능하게 날짜 데이터 변환
df['Date'] = pd.to_datetime(df['Date'])

# 2021년 이후 데이터 필터링
df_after_2021 = df[df['Date'] >= '2021-01-01']

# 데이트별 종가 데이터 출력
plt.plot(df_after_2021['Date'], df_after_2021['Close'])

plt.title('NFLX Close Price')

plt.xlabel('Date')

plt.ylabel('Close Price')

plt.xticks(rotation=45)

plt.show()
```

## C. 데이터 분석 - 2021년 이후 최고, 최저 종가 출력하기

```python
# CSV 파일 경로
csv_path = 'NFLX.csv'

# CSV 파일 읽어오기 (첫 번째, 마지막 열 제외)
df = pd.read_csv(csv_path, usecols=range(0, 5))

# 필터링 가능하게 날짜 데이터 변환
df['Date'] = pd.to_datetime(df['Date'])

# 2021년 이후 데이터 필터링
df_after_2021 = df[df['Date'] >= '2021-01-01']

max_price = max(df_after_2021['Close'])

min_price = min(df_after_2021['Close'])

print('최고 종가:', max_price)
print('최저 종가:', min_price)
```

## D. 데이터 분석 - 2021년 이후 월 별 평균 종가 출력하기

```python
# CSV 파일 경로
csv_path = 'NFLX.csv'

# CSV 파일 읽어오기 (첫 번째, 마지막 열 제외)
df = pd.read_csv(csv_path, usecols=range(0, 5))

# 필터링 가능하게 날짜 데이터 변환
df['Date'] = pd.to_datetime(df['Date'])

# 2021년 이후 데이터 필터링
df_after_2021 = df[df['Date'] >= '2021-01-01']

# 월 별 평균 종가 필터링
df_monthly = pd.DataFrame(df_after_2021.groupby(df_after_2021['Date'].dt.strftime('%Y-%m')).mean())

# 출력
plt.plot(df_monthly['Date'], df_monthly['Close'])

plt.title('Monthly Average Close Price')

plt.xlabel('Date')

plt.ylabel('Average Close Price')

plt.xticks(rotation=45)

plt.show()
```

## E. 데이터 시각화 - 2022년 이후 최고, 최저, 종가 시각화하기

```python
# CSV 파일 경로
csv_path = 'NFLX.csv'

# CSV 파일 읽어오기 (첫 번째, 마지막 열 제외)
df = pd.read_csv(csv_path, usecols=range(0, 5))

# 필터링 가능하게 날짜 데이터 변환
df['Date'] = pd.to_datetime(df['Date'])

# 2022년 이후 데이터 필터링
df_after_2022 = df[df['Date'] >= '2022-01-01']

# 최고, 최저, 종가 데이터 출력
plt.plot(df_after_2022['Date'], df_after_2022['High'], label="High")
plt.plot(df_after_2022['Date'], df_after_2022['Low'], label="Low")
plt.plot(df_after_2022['Date'], df_after_2022['Close'], label="Close")

plt.title('High, Low, and Close Prices since January 2022')

plt.xlabel('Date')

plt.ylabel('Price')

plt.xticks(rotation=45)

plt.legend()

plt.show()
```
